- Três lados formam um triângulo quando a soma de quaisquer dois lados for maior que o terceiro;
- Triângulo Equilátero: três lados iguais;
- Triângulo Isósceles: quaisquer dois lados iguais;

A partir do valor da variação, deve ser enviada às seguintes sugestões:
- Para variação acima de 20%: bonificação para o time de vendas.
- Para variação entre 2% e 20%: pequena bonificação para time de vendas.
- Para variação entre 2% e -10%: planejamento de políticas de incentivo às vendas.
- Para bonificações abaixo de -10%: corte de gastos.


````python
# Coleta os dados de vendas anuais
vendas_2022 = float(input('Digite o valor de vendas de 2022: '))
vendas_2023 = float(input('Digite o valor de vendas de 2023: '))

# Calcula a variação percentual
if vendas_2022 == 0:
    print('Não é possível calcular a variação percentual com vendas de 2022 iguais a zero.')
else:
    variacao_percentual = ((vendas_2023 - vendas_2022) / vendas_2022) * 100
    print(f'Variação percentual: {variacao_percentual:.2f}%')

    # Envia sugestões baseadas na variação
    if variacao_percentual > 20:
        print('Sugestão: Bonificação para o time de vendas.')
    elif 2 <= variacao_percentual <= 20:
        print('Sugestão: Pequena bonificação para o time de vendas.')
    elif -10 <= variacao_percentual < 2:
        print('Sugestão: Planejamento de políticas de incentivo às vendas.')
    else: # variacao_percentual < -10
        print('Sugestão: Corte de gastos.')
````

#### Questão 3
Para tratar uma quantidade de 15 dados de avaliações de pessoas usuárias de um serviço da empresa, precisamos verificar se as notas são válidas. Então, escreva um programa que vai receber a nota de 0 a 5 de todos os dados e verificar se é um valor válido. Caso seja inserido uma nota acima de 5 ou abaixo de 0, repita até que a pessoa usuária insira um valor válido.

````python
# Vamos repetir o processo 15 vezes (uma para cada usuário)
for i in range(15):

    # Pede a nota inicial
    nota = float(input(f"Digite a nota do usuário {i+1}: "))

    # VALIDAÇÃO: Enquanto a nota for menor que 0 OU maior que 5...
    while (nota < 0) or (nota > 5):
        print("❌ Nota Inválida! A nota deve ser entre 0 e 5.")
        # Pede a nota novamente (e atualiza a variável 'nota')
        nota = float(input(f"Tente novamente a nota do usuário {i+1}: "))

    # Se saiu do while, significa que a nota é válida
    print("✅ Nota Válida computada.")

print("Fim das avaliações!")
````

#### Questão 5
Escreva um programa que calcule o fatorial de um número inteiro fornecido pela pessoa usuária. O fatorial de um número inteiro é a multiplicação desse número por todos os seus antecessores até o número 1. Por exemplo, o fatorial de 5 é 5 x 4 x 3 x 2 x 1 = 120.

````python
# 1. Entrada de dados
numero = int(input("Digite um número inteiro: "))

# 2. Peças do quebra-cabeça
valoracumulado = 1
contador = numero

print(f"\n⚙️ Iniciando o cálculo de {numero}! (fatorial)")
print("-" * 40)

# 3. O Laço de Repetição
while contador > 1:
    # Guardamos o valor antes da conta apenas para exibir no print depois
    valor_anterior = valoracumulado

    # Realizamos a multiplicação
    valoracumulado = valoracumulado * contador

    # EXIBIÇÃO DO RACIOCÍNIO:
    print(f"🔄 No degrau {contador}:")
    print(f"   Conta: {valor_anterior} (acumulado) x {contador} (contador)")
    print(f"   Novo valor na caixa: {valoracumulado}")
    print("-" * 40)

    # Diminuímos o contador para descer um degrau
    contador = contador - 1

# 4. Exibição do resultado final
print(f"✅ Fim do cálculo! O fatorial de {numero} é: {valoracumulado}")
````

#### Questão 7
Os números primos possuem várias aplicações dentro da Ciência de Dados, por exemplo, na criptografia e segurança. Um número primo é aquele que é divisível apenas por um e por ele mesmo. Faça um programa que peça um número inteiro e determine se ele é ou não um número primo.

````python
numero = int(input('Digite um número inteiro: '))

if numero <= 1:
  print(f'O número {numero} não é primo.')
elif numero == 2:
  print(f'O número {numero} é primo.')
else:
  is_primo = True
  divisor = 2
  # Laço de repetição mais simples, verifica divisores até o número - 1
  while divisor < numero:
    if numero % divisor == 0:
      is_primo = False
      break # Sai do laço se encontrar um divisor
    divisor += 1

  if is_primo:
    print(f'O número {numero} é primo.')
  else:
    print(f'O número {numero} não é primo.')
````

#### Questão 9
Em uma eleição para gerência em uma empresa com 20 funcionários, existem quatro candidatos. Escreva um programa que calcule o vencedor da eleição. A votação ocorreu da seguinte maneira:
- Cada funcionário votou em um dos quatro candidatos (representados pelos números 1, 2, 3 e 4).
- Também foram contabilizados os votos nulos (representado pelo número 5) e os votos em branco (representado pelo número 6).

Ao final da votação, o programa deve exibir o total de votos para cada candidato, o número de votos nulos e o número de votos em branco. Além disso, deve calcular e exibir a porcentagem de votos nulos em relação ao total de votos e a porcentagem de votos em branco em relação ao total de votos.

````python
num_funcionarios = 20

votos_candidato1 = 0
votos_candidato2 = 0
votos_candidato3 = 0
votos_candidato4 = 0
votos_nulos = 0
votos_em_branco = 0

print('--- Eleição para Gerência ---')
print('Candidatos: 1, 2, 3, 4')
print('Opções: 5 (Nulo), 6 (Em Branco)')

for i in range(num_funcionarios):
    voto = 0 # Inicializa voto para entrar no loop
    while not (1 <= voto <= 6):
        voto_str = input(f'Funcionário {i + 1}, digite seu voto (1-6): ')
        # A validação de tipo de entrada não usa try-except, assume que é um número.
        # Se o usuário digitar algo não numérico, o programa pode falhar.
        if voto_str.isdigit(): # Verifica se a entrada é composta apenas por dígitos
            voto = int(voto_str)
            if not (1 <= voto <= 6):
                print('Voto inválido. Por favor, digite um número entre 1 e 6.')
        else:
            print('Entrada inválida. Por favor, digite um número inteiro.')
            voto = 0 # Reinicia voto para manter o loop

    if voto == 1:
        votos_candidato1 += 1
    elif voto == 2:
        votos_candidato2 += 1
    elif voto == 3:
        votos_candidato3 += 1
    elif voto == 4:
        votos_candidato4 += 1
    elif voto == 5:
        votos_nulos += 1
    else: # voto == 6
        votos_em_branco += 1

total_votos = num_funcionarios

print('\n--- Resultados da Eleição ---')
print(f'Candidato 1: {votos_candidato1} votos')
print(f'Candidato 2: {votos_candidato2} votos')
print(f'Candidato 3: {votos_candidato3} votos')
print(f'Candidato 4: {votos_candidato4} votos')
print(f'Votos Nulos: {votos_nulos} votos')
print(f'Votos em Branco: {votos_em_branco} votos')

# Cálculo das porcentagens
porcentagem_nulos = (votos_nulos / total_votos) * 100 if total_votos > 0 else 0
porcentagem_em_branco = (votos_em_branco / total_votos) * 100 if total_votos > 0 else 0

print(f'Porcentagem de votos nulos: {porcentagem_nulos:.2f}%')
print(f'Porcentagem de votos em branco: {porcentagem_em_branco:.2f}%')

# Determinar o vencedor
vencedor = None
max_votos = -1

# Criando uma lista simples de candidatos e seus votos para facilitar a comparação
candidatos_votos = [
    (1, votos_candidato1),
    (2, votos_candidato2),
    (3, votos_candidato3),
    (4, votos_candidato4)
]

# Encontra o máximo de votos entre os candidatos
# Usa uma flag para detectar empates
possiveis_vencedores = []

for cand_num, votos_atual in candidatos_votos:
    if votos_atual > max_votos:
        max_votos = votos_atual
        possiveis_vencedores = [cand_num] # Novo maior, reseta a lista
    elif votos_atual == max_votos and max_votos > 0: # Adiciona em caso de empate
        possiveis_vencedores.append(cand_num)

if len(possiveis_vencedores) > 1:
    print(f'\nResultado: Empate entre os Candidatos {', '.join(map(str, possiveis_vencedores))} com {max_votos} votos!')
elif len(possiveis_vencedores) == 1:
    print(f'\nO vencedor da eleição é o Candidato {possiveis_vencedores[0]} com {max_votos} votos!')
else:
    print('\nNão houve votos válidos para nenhum candidato.')
````
